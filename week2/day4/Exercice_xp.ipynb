## Exercise 1: Understanding Data Visualization

**Task 1:** Why is data visualization important?

Data visualization helps transform raw data into charts and graphs that are easier to understand. It allows analysts to identify trends, patterns, outliers, and relationships, making data-driven decision-making more effective.

**Task 2:** Purpose of a Line Graph

A line graph is used to visualize changes over time or across an ordered sequence. It helps reveal trends, increases, decreases, and fluctuations in data.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import zipfile
```

## Exercise 2: Line Plot for Temperature Variation

```python
days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
temperatures = [72, 74, 76, 80, 82, 78, 75]

plt.figure(figsize=(8,5))
plt.plot(days, temperatures, marker='o')
plt.xlabel('Day')
plt.ylabel('Temperature (°F)')
plt.title('Temperature Variation Over a Week')
plt.grid(True)
plt.show()
```

## Exercise 3: Monthly Sales Bar Chart

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [5000, 5500, 6200, 7000, 7500]

plt.figure(figsize=(8,5))
plt.bar(months, sales)
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.title('Monthly Sales Data')
plt.show()
```

## Load the Student Mental Health Dataset

```python
zip_path = 'Student Mental health.zip'
with zipfile.ZipFile(zip_path, 'r') as z:
    z.extractall('.')

df = pd.read_csv('Student Mental health.csv')
df.head()
```

## Exercise 4: Histogram of CGPA Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='What is your CGPA?', color='skyblue')
plt.title('Distribution of Students CGPA')
plt.xticks(rotation=45)
plt.show()
```

## Exercise 5: Anxiety Levels Across Genders

```python
anxiety_data = pd.crosstab(df['Choose your gender'], df['Do you have Anxiety?'], normalize='index').reset_index()
anxiety_long = anxiety_data.melt(id_vars='Choose your gender', var_name='Anxiety', value_name='Proportion')

plt.figure(figsize=(8,5))
sns.barplot(data=anxiety_long, x='Choose your gender', y='Proportion', hue='Anxiety', palette='Set2')
plt.title('Proportion of Anxiety Across Genders')
plt.show()
```

## Exercise 6: Relationship Between Age and Panic Attacks

```python
df['Panic_Attack_Numeric'] = df['Do you have Panic attack?'].map({'Yes':1, 'No':0})

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Age', y='Panic_Attack_Numeric', style='Choose your gender')
plt.xlabel('Age')
plt.ylabel('Panic Attack (1=Yes, 0=No)')
plt.title('Age vs Panic Attacks')
plt.show()
```
