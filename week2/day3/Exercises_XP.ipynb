## Exercise 2: Descriptive Statistics
Given a sample dataset, calculate the mean, median, variance, and standard deviation using SciPy.

```python
import numpy as np
from scipy import stats

data = [12, 15, 13, 12, 18, 20, 22, 21]

mean_val = np.mean(data)
median_val = np.median(data)
variance_val = np.var(data, ddof=1)   # sample variance
std_dev_val = np.std(data, ddof=1)     # sample standard deviation

print("Data:", data)
print("Mean:", mean_val)
print("Median:", median_val)
print("Variance (sample):", variance_val)
print("Standard Deviation (sample):", std_dev_val)
```

## Exercise 3: Understanding Distributions
Generate a normal distribution using SciPy with a mean of 50 and a standard deviation of 10, then plot it.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu, sigma = 50, 10
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label=f"Normal Distribution (μ={mu}, σ={sigma})")
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

## Exercise 4: T-Test Application
Perform a t-test on two sets of randomly generated data.

```python
import numpy as np
from scipy import stats

np.random.seed(42)
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_stat, p_value = stats.ttest_ind(data1, data2, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: the two groups are significantly different.")
else:
    print("Fail to reject the null hypothesis: no significant difference detected.")
```

## Exercise 5: Linear Regression Analysis
Apply linear regression to housing prices and interpret the results.

```python
import numpy as np
from scipy import stats

house_sizes = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

predicted_price_90 = slope * 90 + intercept

print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"R-squared: {r_value**2:.4f}")
print(f"Predicted price for 90 m²: {predicted_price_90:.2f}")
print()
print("Interpretation:")
print(f"Each additional 1 m² is associated with an increase of about {slope:.2f} currency units in price, on average.")
```

## Exercise 6: Understanding ANOVA
Test whether three fertilizers have significantly different effects on plant growth.

```python
from scipy import stats

fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

f_stat, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print("F-value:", f_stat)
print("P-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: the fertilizers have significantly different effects.")
else:
    print("Fail to reject the null hypothesis: no significant difference detected.")

print()
print("If the P-value were greater than 0.05, we would fail to reject the null hypothesis,")
print("meaning there would be insufficient evidence that the fertilizers differ significantly.")
```

## Exercise 7: Probability Distributions (Optional)
Work with a binomial distribution: calculate the probability of getting exactly 5 heads in 10 coin flips.

```python
from scipy.stats import binom

n = 10
p = 0.5
k = 5

prob_exactly_5 = binom.pmf(k, n, p)

print(f"Probability of exactly {k} heads in {n} flips: {prob_exactly_5:.6f}")
```

## Exercise 8: Correlation Coefficients (Optional)
Calculate the Pearson and Spearman correlation coefficients between age and income.

```python
import pandas as pd
from scipy import stats

data = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

pearson_corr, pearson_p = stats.pearsonr(data['age'], data['income'])
spearman_corr, spearman_p = stats.spearmanr(data['age'], data['income'])

print("Dataset:")
print(data)
print()
print(f"Pearson correlation: {pearson_corr:.6f} (p-value: {pearson_p:.6f})")
print(f"Spearman correlation: {spearman_corr:.6f} (p-value: {spearman_p:.6f})")
```
