# NumPy Exercises

This notebook contains 10 simple NumPy exercises with code and outputs.

## Exercise 1: Array Creation and Manipulation

```python
import numpy as np

arr1 = np.arange(10)
arr1
```

## Exercise 2: Type Conversion and Array Operations

```python
arr2 = np.array([3.14, 2.17, 0, 1, 2], dtype=int)
arr2
```

## Exercise 3: Working with Multi-Dimensional Arrays

```python
arr3 = np.arange(1, 10).reshape(3, 3)
arr3
```

## Exercise 4: Creating Multi-Dimensional Array with Random Numbers

```python
np.random.seed(42)
arr4 = np.random.rand(4, 5).round(2)
arr4
```

## Exercise 5: Indexing Arrays

```python
arr5 = np.array([[21, 22, 23, 22, 22],
                 [20, 21, 22, 23, 24],
                 [21, 22, 23, 22, 22]])

second_row = arr5[1]
second_row
```

## Exercise 6: Reversing Elements

```python
arr6 = np.arange(10)
arr6[::-1]
```

## Exercise 7: Identity Matrix

```python
arr7 = np.identity(4)
arr7
```

## Exercise 8: Simple Aggregate Functions

```python
arr8 = np.arange(10)
array_sum = arr8.sum()
array_avg = arr8.mean()
print(f"Sum: {array_sum}, Average: {array_avg}")
```

## Exercise 9: Create Array and Change its Structure

```python
arr9 = np.arange(1, 21).reshape(4, 5)
arr9
```

## Exercise 10: Conditional Selection of Values

```python
arr10 = np.arange(1, 11)
arr10[arr10 % 2 == 1]
```
