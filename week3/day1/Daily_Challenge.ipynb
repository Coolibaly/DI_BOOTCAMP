
# Daily Challenge: NumPy, Pandas, and Matplotlib Integration

## Objective
Analyze and visualize synthetic temperature data for 10 cities across 12 months using NumPy, Pandas, and Matplotlib.


```

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
```

## 1. Data Preparation

```

cities = [
    'Abidjan', 'Paris', 'London', 'New York', 'Tokyo',
    'Sydney', 'Cairo', 'Lagos', 'Toronto', 'Berlin'
]

months = [
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
]

temperature_data = np.random.uniform(
    low=-5,
    high=35,
    size=(10, 12)
)

df = pd.DataFrame(
    temperature_data,
    index=cities,
    columns=months
)

df.round(2)
```

## 2. Data Analysis

```

annual_avg = df.mean(axis=1)

highest_city = annual_avg.idxmax()
lowest_city = annual_avg.idxmin()

print("Annual Average Temperatures")
print(annual_avg.round(2))

print("\nCity with Highest Average Temperature:", highest_city)
print("Average:", round(annual_avg.max(), 2), "°C")

print("\nCity with Lowest Average Temperature:", lowest_city)
print("Average:", round(annual_avg.min(), 2), "°C")
```

## 3. Data Visualization

```

plt.figure(figsize=(12, 6))

for city in df.index:
    plt.plot(months, df.loc[city], marker='o', label=city)

plt.title('Monthly Temperature Trends by City')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```

plt.figure(figsize=(10, 5))

annual_avg.sort_values().plot(kind='bar')

plt.title('Annual Average Temperature by City')
plt.xlabel('City')
plt.ylabel('Average Temperature (°C)')
plt.tight_layout()
plt.show()
```


## Findings

This analysis generated synthetic weather data for 10 cities over 12 months.

Key outputs:
- Computed annual average temperatures for each city.
- Identified the city with the highest annual average temperature.
- Identified the city with the lowest annual average temperature.
- Visualized monthly temperature trends using a line chart.
- Compared annual averages using a bar chart.

Interesting trends can be observed by examining seasonal fluctuations and comparing average temperatures across cities.
