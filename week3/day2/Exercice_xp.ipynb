## Exercise 1: Matrix Operations

We create a 3×3 matrix, then compute:

1. Its determinant
2. Its inverse

```
# 3x3 matrix
A = np.array([
    [2, 1, 3],
    [1, 0, 2],
    [4, 1, 8]
])

print("Matrix A:")
print(A)

det_A = np.linalg.det(A)
print("\nDeterminant of A:", det_A)

# Inverse exists only if determinant is non-zero
if not np.isclose(det_A, 0):
    inv_A = np.linalg.inv(A)
    print("\nInverse of A:")
    print(inv_A)

    # Check: A @ inv_A should be close to identity
    print("\nA @ inv(A):")
    print(np.round(A @ inv_A, 6))
else:
    print("\nThis matrix is not invertible because its determinant is 0.")
```

## Exercise 6: Basic Hypothesis Testing

We test whether a new training program improves productivity.

### Hypotheses
- **H0**: The training program does not improve productivity.
- **H1**: The training program improves productivity.

We use the mean difference and a simple t-statistic computed with NumPy.

```
# Productivity scores before training
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Productivity scores after training (slight improvement)
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

# Paired differences
diff = productivity_after - productivity_before

mean_diff = np.mean(diff)
std_diff = np.std(diff, ddof=1)
n = len(diff)
se_diff = std_diff / np.sqrt(n)

# Simple paired t-statistic
t_stat = mean_diff / se_diff

print("Mean before training:", np.mean(productivity_before))
print("Mean after training:", np.mean(productivity_after))
print("Mean improvement:", mean_diff)
print("Standard deviation of improvement:", std_diff)
print("t-statistic:", t_stat)

# Simple interpretation using a one-sided 5% threshold (approx. normal/t critical ~ 1.645)
if t_stat > 1.645:
    print("\nConclusion: Reject H0. The training program appears to improve productivity.")
else:
    print("\nConclusion: Fail to reject H0. Not enough evidence of improvement.")
```

## Exercise 8: Time Series Data Manipulation

We generate daily time series data for 2023, then slice it by quarters:

- January to March
- April to June
- July to September
- October to December

```
date_range_2023 = pd.date_range("2023-01-01", "2023-12-31", freq="D")
time_series = pd.Series(np.random.randn(len(date_range_2023)).cumsum(), index=date_range_2023)

print("Full time series sample:")
print(time_series.head())

jan_mar = time_series["2023-01":"2023-03"]
apr_jun = time_series["2023-04":"2023-06"]
jul_sep = time_series["2023-07":"2023-09"]
oct_dec = time_series["2023-10":"2023-12"]

print("\nJanuary to March:")
print(jan_mar.head())

print("\nApril to June:")
print(apr_jun.head())

print("\nJuly to September:")
print(jul_sep.head())

print("\nOctober to December:")
print(oct_dec.head())
```

## Exercise 9: Data Conversion

Convert a NumPy array to a Pandas DataFrame and then back to a NumPy array.

```
array_np = np.array([[1, 2, 3], [4, 5, 6]])

df_from_np = pd.DataFrame(array_np, columns=["X", "Y", "Z"])
array_back = df_from_np.to_numpy()

print("Original NumPy array:")
print(array_np)

print("\nDataFrame from NumPy array:")
print(df_from_np)

print("\nBack to NumPy array:")
print(array_back)
```
